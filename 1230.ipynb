````python
#題目:符號0 從最右邊開始印十行(層)每層遞增,空白也要印X
#方法1
layer =10
left,right ="X", "O"
i=0
while i < layer:
    leftcount = layer-i-1
    rightcount = i+1
    print(left * leftcount +
       right * rightcount)
    i = i +1
````

````python
#題目:符號倒金字塔每層遞減,符號為O與X循環
方法1
layer = 5
i = 0
while i < layer:
    # 奇數層: i = 0 2 4 6 8
    if i % 2 == 0:
        left, mid, right = " ", "o", " "
    # 偶數層(% 2 == 1)
    else:
        left, mid, right = " ", "x", " "
    lcount = i
    mcount = (2 * layer - 1) - 2 * i
    rcount = i
    print(left * lcount +
       mid * mcount +
       right * rcount)
    i = i + 1
````

````python
#題目:符號倒金字塔每層遞減,符號為4個Unicode接續循環
#方法1
symbols = ["\u2660", "\u2661", "\u2662", "\u2663"]
layer = 10
i = 0

while i < layer:
    # 符號 中間符號每四個一循環-> %取餘
    lsymbol = " "
    msymbol = symbols[i%len(symbols)]
    rsymbol = " "
    # 數目
    lcount = i
    mcount = (2 * layer - 1) - 2 * i
    rcount = i
    print(lsymbol * lcount +
       msymbol * mcount +
       rsymbol * rcount)
    i = i + 1
````

````python
#題目:給你兩個數字a和b(a<b),請幫我算出a/b的十位小數(除的盡就都是0)
a, b = 2, 7
ans = "0."

i = 0
while i < 10:
    #補0 讓2 -> 20
    a10 = a * 10
    #取商 20 // 7 -> 2 (轉字串接在 "0"後)
    ans = ans + str(a10 // b)
    #取餘 20 % 7 (餘數為新的a,跑下一輪迴圈)
    a = a10 % b
    print(ans)
    i = i + 1
````

````python
#題目:英文大小寫,統計出來s裡面有多少個小寫 多少個大寫
#參照ASCII code對照表 英文字母不論大小寫的十進位數字都是遞增
#方法1:

s = "asdfsDfgvv   ,, WqrRrwqr. wQr123123rqw5tAfg"
uppercase, lowercase = 0, 0
i = 0
while i < len(s):
    #先把英文字母轉成十進位數字n
    n = ord(s[i])
    #判斷n是在英文的大寫區間還是小寫區間
    if ord("A") <= n <= ord("Z"):
        uppercase = uppercase + 1
    elif ord("a") <= n <= ord("z"):
        lowercase = lowercase + 1
    i = i + 1
print(uppercase, lowercase)
print(ord("A"))
````

````python
#題目:統計大小寫字母每一個字母各自有幾個
#方法1

s = "asdfsDfgvv   ,, WqrRrwqr. wQr123123rqw5tAfg"
uppercase = [0] * 26
lowercase = [0] * 26
i = 0
while i < len(s):
    #先把英文字母轉成十進位數字n
    n = ord(s[i])
    #判斷n是在英文的大寫區間還是小寫區間
    if ord("A") <= n <= ord("Z"):
        # 65 - 90 -> 0 - 25 把ASCII轉換成對應大小寫(uppercase,lowercase)清單位置
        n = n - ord("A")
        uppercase[n] = uppercase[n] + 1
    elif ord("a") <= n <= ord("z"):
        # 97 - 122 -> 0 - 25
        n = n - ord("a")
        lowercase[n] = lowercase[n] + 1
    i = i + 1
print(uppercase)
print(lowercase)

#題目: A-Z印出26行顯示個別大小寫數量(把上題結果從橫式轉直式)
# A: 5  a: 3 ?????????????????????
# B: 2  b: 4 ?????????????????????
# chr(97) -> "a" chr(65) -> "A"
i = 0
while i < len(uppercase):
    c1, c2 = chr(ord("A") + i ), chr(ord("a") + i)
    n1, n2 = uppercase[i], lowercase[i]
    #c1是大寫c2是小寫,n1與n2是其對應的數量
    print(c1, n1, "/", c2, n2)
    i = i + 1
````

````python
#方法2

alphabets = "asdfsDfgvv   ,, WqrRrwqr. wQr123123rqw5tAfg"
uppercase = [0] * 26
lowercase = [0] * 26

for a in alphabets:
    #先把英文字母轉成十進位數字n
    n = ord(a)
    #判斷n是在英文的大寫區間還是小寫區間
    if ord("A") <= n <= ord("Z"):
        # 65 - 90 -> 0 - 25 把ASCII轉換成對應大小寫(uppercase,lowercase)清單位置
        n = n - ord("A")
        uppercase[n] = uppercase[n] + 1
    elif ord("a") <= n <= ord("z"):
        # 97 - 122 -> 0 - 25
        n = n - ord("a")
        lowercase[n] = lowercase[n] + 1
    i = i + 1
print(uppercase)
print(lowercase)

#題目: A-Z印出26行顯示個別大小寫數量(把上題結果從橫式轉直式)
# A: 5  a: 3
# B: 2  b: 4
# chr(97) -> "a" chr(65) -> "A"
i = 0
while i < len(uppercase):
    c1, c2 = chr(ord("A") + i ), chr(ord("a") + i)
    n1, n2 = uppercase[i], lowercase[i]
    #c1是大寫c2是小寫,n1與n2是其對應的數量
    print(c1, n1, "/", c2, n2)
    i = i + 1
````

````python
#題目: 求半徑為1的圓面積
#方法1

# x的最左邊座標和x的最右邊座標
x_left, x_right = -1, 1
# 分成幾份(本題假設以100份作計算,下方數字僅最後用來變動測試證明分得越細,面積答案越精準)
part = 100
# 每一part的x間隔多長
interval = (x_right - x_left) / part

# 總面積
# 積分法: 把每一part的面積(小長方形)加總起來
total = 0
i = 0
while i < part:
    # 每一part的左x座標: -1 + 0.02 * i
    x = x_left + interval * i
    # 算出這裡的y座標多長(H): 2 * (1 - x ^ 2) ^ 0.5
    y = 2 * (1 - x ** 2) ** 0.5
    # 每一part面積
    area = interval * y
    # 把每一part累積進去
    total = total + area
    i = i + 1
print("積分法估算面積:", total)
````

````python
#方法2

# x的最左邊座標和x的最右邊座標
x_left, x_right = -1, 1
# 分成幾份(本題假設以100份作計算,下方數字僅最後用來測試證明分得越細,面積答案越精準)
part = 100
# 每一part的x間隔多長
interval = (x_right - x_left) / part

# 總面積
# 積分法: 把每一part的面積(當成極小長方形)加總起來
total = 0

for i in range(100):
    # 每一part的左x座標: -1 + 0.02 * i
    x = x_left + interval * i
    # 算出這裡的y座標多長(H): 2 * (1 - x ^ 2) ^ 0.5
    y = 2 * (1 - x ** 2) ** 0.5
    # 每一part面積
    area = interval * y
    # 把每一part累積進去
    total = total + area

print("積分法估算面積:", total)
````
