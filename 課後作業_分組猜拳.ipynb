```python
import random

# my = int(input("0-5:"))
# com = random.randint(0, 5)
my = int(input("請輸入0-5 :"))
com = random.randint(0,5)

trans = ["剪刀a", "石頭a", "布a",
      "剪刀b", "石頭b", "布b"]

print("我出的:", trans[my])
print("電腦的:", trans[com])

# 第一步:先確定他們是否同組
# 0 // 3: 0
# 1 // 3: 0
# 2 // 3: 0
# 3 // 3: 1
# 4 // 3: 1
# 5 // 3: 1
# 6 // 3: 2
# 取商: 分組 取餘: 循環
mygroup = my // 3
comgroup = com // 3
    # 情況1: my, com都是0~2(0~2)
if mygroup == comgroup == 0:
    if my == (com + 1) % 3:
        print("Win")
    elif com == (my + 1) % 3:
        print("Lose")
    else:
        print("Even")
    # 情況2: my, com都是3~5(0~2)回到0~2(餘數
elif mygroup == comgroup == 1:
    if my % 3 == (com % 3 + 1 ) % 3:
        print("Win")
    elif com % 3 == (my % 3 + 1) % 3:
        print("Lose")
    else:
        print("Even")
    # 情況3: 不同組為平手
else:
    print("平手(不同組)")
```
