````python
#題目:猜拳贏三把才獲勝
#屬於"其他條件迴圈"有兩種寫法
#1.While+條件() win < 3 and lose < 3:
#2.While True(先不加條件) 最後收尾再下IF判斷是否結束+break(相對IF要縮排)

#隨機整數 -1:輸/ 0:平手 / 1:贏
import random
trans = ["輸(L)","平手(E)","贏(W)"]
win,even,lose = 0,0,0
while True:
    result = random.randint(-1,1)
    #-1,0,1 對應trans字串位置是0,1,2 所以要+1
    print(trans[result+1])

    if result == -1:
        lose = lose + 1
    elif result == 1:
        win = win + 1
    elif result == 0:
        even = even +1
    if lose == 3:
      print("L")
      break
    elif win == 3:
      print("W")
      break
````

````python
# 平手(E)不影響勝負,也不影響迴圈結束與否,所以可略
import random
win,lose = 0,0
while True:
    result = random.randint(-1,1)
    #相對上一種寫法沒有對應到字串文字所以沒有+1 直接print(result)
    print(result)

    if result == -1:
        lose = lose + 1
    elif result == 1:
        win = win + 1

    if lose == 3:
      print("L")
      break
    elif win == 3:
      print("W")
      break
````

````python
#Q1給你兩個數字a和b(a<b),請幫我算出a/b的十位小數(除的盡就都是0)
a, b = 2, 7
ans = "0."

i = 0
while i < 10:
    #補0 讓2 -> 20
    a10 = a * 10
    #取商 20 // 7 -> 2 (轉字串接在 "0"後)
    ans = ans + str(a10 // b)
    #取餘 20 % 7 (餘數為新的a,跑下一輪迴圈)
    a = a10 % b
    print(ans)
    i = i + 1
````

````python
# n = 19800229
# n = 20000229
# n = 30000229
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

def is_valid_date(n):
    year = n // 10000
    last4 = n % 10000
    day = last4 % 100
    month = last4 // 100
    # print(year, month, day)

    max_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if 1 <= month <= 12:
        month_day = max_days[month-1]
        # 考慮閏年
        if month == 2 and is_leap_year(year) == True:
            month_day = month_day + 1

        if 1 <= day <= month_day:
            # print("合理")
            return True
        else:
            # print("不合理日期(天)")
            return False
    else:
        # print("不合理日期(月)")
        return False

print(is_valid_date(19800229))
print(is_valid_date(20000229))
print(is_valid_date(30000229))

#目標: 給妳隨便一個日期, 幫我往後出10個完全平方日
#Key: 某個日期是不是合理!!!

# 正面: 把真實的日期拿出來, 檢查她可以不可以被開根號
# 反面: 先拿值出來, 平方, 看這個平方過後的東西可不可以當日期(o)
# 選擇反面思考寫法較簡單
start = 4500
count = 0
while True:
    date = start ** 2
    if is_valid_date(date) == True:
        print(start, date, "是個完全平方日")
        count = count + 1
    #start不能縮排到IF下一層,否則當條件不成立時start不會增加，迴圈會繼續檢查相同的數字。
    start = start + 1
    if count == 10:
        break
````
